```
input_shape = (128, 128, 1)
```

```
model = Sequential()

model.add(Conv2D(64, kernel_size=(3, 3), activation='relu',
                input_shape=input_shape))

model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))

model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(256, kernel_size=(3, 3), activation='relu'))

model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(512, kernel_size=(3, 3), activation='relu',
          kernel_regularizer=regularizers.l2(0.01)))

model.add(Flatten())

model.add(Dropout(0.5))

model.add(Dense(128, activation='relu',
          kernel_regularizer=regularizers.l2(0.01)))

model.add(Dropout(0.2))

model.add(Dense(10, activation='softmax'))
```

```
model.compile(loss='categorical_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
```

```
genres = ['blues', 'classical', 'country', 'disco', 'hiphop',
          'jazz', 'metal', 'pop', 'reggae', 'rock']

ys = np.eye(10)
```

```
!cd spec2 && rm .DS_Store 
```

```
y = {'train':[], 'test':[], 'valid':[]}

for direc in os.listdir('spec2'):
  for file in os.listdir('spec2/'+direc):
    for i, genre in enumerate(genres):  
      if file.startswith(genre):
        y[direc].append(ys[i])
```

```
from keras.preprocessing.image import load_img, img_to_array
```

```
batch_size = 32
```

```
train_data = []

for training_file in os.listdir('spec2/train/'):
  img = load_img('spec2/train/'+training_file,
                 target_size=(128, 128),
                 grayscale=True)
  train_data.append(img_to_array(img))

train_data = np.array(train_data)
```

```
valid_data = []

for valid_file in os.listdir('spec2/valid/'):
  img = load_img('spec2/valid/'+valid_file,
                 target_size=(128, 128),
                 grayscale=True)
  valid_data.append(img_to_array(img))
  
valid_data = np.array(valid_data)
```

```
test_data = []

for test_file in os.listdir('spec2/test/'):
  img = load_img('spec2/test/'+test_file,
                 target_size=(128, 128),
                 grayscale=True)
  test_data.append(img_to_array(img))
  
test_data = np.array(test_data)
```

```
train_data /= 255
test_data /= 255
valid_data /= 255
```

```
model.fit(train_data, np.array(y['train']),
         epochs=50,
         batch_size=batch_size,
         validation_data=(valid_data, np.array(y['valid'])),
         verbose=1)
```

```
model.save_weights('audio_model_3.h5')
```

```
predictions = model.predict(test_data, batch_size=batch_size, verbose=1)
```

```
is_correct = 0

for i in range(432):
  if np.argmax(y['test'][i]) == np.argmax(predictions[i]):
    is_correct += 1
```

```
accuracy = is_correct/432
```
